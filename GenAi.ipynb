## What is Generative AI?

Generative Artificial Intelligence (AI) refers to algorithms that can **create new content**, such as text, images, music, or even code. These models learn patterns from existing data and use this knowledge to generate **novel outputs** that resemble the training data.

Generative AI has a wide range of applications, from writing stories and generating poetry to creating realistic images and composing music.

#**Text Generation with Pre-trained Models**
We'll use a pre-trained language model to generate text based on our input prompts. Specifically, we'll use GPT-2, a popular model developed by OpenAI.

##Load the Text Generation Pipeline


```python
# Load the text generation pipeline with GPT-2
generator = pipeline('text-generation', model='gpt2')
```

##Generate Text from a Prompt
Let's start by generating text from a simple prompt.



```python
# Define a prompt
prompt = "tell me a joke"

# Generate text
generated_text = generator(prompt, max_length=50, num_return_sequences=1)

# Display the generated text
print(generated_text[0]['generated_text'])
```

#**Interactive Story Generation**
Now it's your turn! Input a starting sentence, and the model will generate the rest.


```python
# Get user input
user_prompt = input("Enter a starting sentence for your story: ")

# Generate text
story = generator(user_prompt, max_length=100, num_return_sequences=1)

# Display the story
print("\nHere is your story:\n")
print(story[0]['generated_text'])
```

#**Experiment with Generation Settings**
You can tweak the generation parameters to see how they affect the output.

max_length: The maximum number of tokens (words or subwords) to generate.

num_return_sequences: How many different outputs to generate.

temperature: Controls randomness; higher values make the output more random.

```python
# Customize generation settings
user_prompt = input("Enter a prompt: ")
max_length = int(input("Enter the maximum length of the generated text (e.g., 50): "))
num_sequences = int(input("Enter the number of sequences to generate (e.g., 3): "))
temperature = float(input("Enter the temperature (e.g., 0.7): "))

# Generate text with custom settings
custom_texts = generator(
    user_prompt,
    max_length=max_length,
    num_return_sequences=num_sequences,
    temperature=temperature,
    do_sample=True,
)

# Display the generated texts
for idx, text in enumerate(custom_texts):
    print(f"\nGenerated Text {idx + 1}:\n")
    print(text['generated_text'])
```

#**Trying Different Models**
Let's try a larger model for potentially better results. We'll use GPT-2 Medium.

```python
# Load GPT-2 Medium
generator_medium = pipeline('text-generation', model='gpt2-medium')
```

```python
# Get user input
medium_prompt = input("Enter a prompt for the larger model: ")

# Generate text
medium_text = generator_medium(medium_prompt, max_length=100, num_return_sequences=1)

# Display the generated text
print("\nGenerated Text with GPT-2 Medium:\n")
print(medium_text[0]['generated_text'])
```

#**Fun with Poetry Generation**
Let's generate a poem! Provide a poetic starting line.

```python
# Poetic prompt
poetry_prompt = input("Enter the first line of a poem: ")

# Generate a poem
poem = generator(
    poetry_prompt,
    max_length=50,
    num_return_sequences=1,
    temperature=1.0,
    do_sample=True,
)

# Display the poem
print("\nHere is your poem:\n")
print(poem[0]['generated_text'])
```

#**Generating Text in a Specific Style**
You can guide the model to write in a particular style by tweaking the prompt.

```python
# Prompt for Shakespearean style
style_prompt = "In fair Verona, where we lay our scene,"

# Generate text
styled_text = generator(
    style_prompt,
    max_length=100,
    num_return_sequences=1,
    temperature=0.8,
    do_sample=True,
)

# Display the styled text
print("\nGenerated Text in Shakespearean Style:\n")
print(styled_text[0]['generated_text'])
```
